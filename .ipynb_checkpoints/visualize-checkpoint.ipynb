```python
# 
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
from wordcloud import WordCloud
import jieba
from collections import Counter
import os

#  data/ 
font_path = 'data/PingFang.ttc'
font_prop = fm.FontProperties(fname=font_path)
plt.rcParams['font.family'] = font_prop.get_name()

# 
os.makedirs("data", exist_ok=True)

#  CSV 
df = pd.read_csv('data/cheergirls_articles_3pages.csv')

# 
def generate_wordcloud(text, output_path):
    font_path = 'data/PingFang.ttc'  # 

    wordcloud = WordCloud(
        font_path=font_path,
        width=800, height=400,
        background_color='white',
        colormap='Blues'
    ).generate(text)

    # 
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.tight_layout()
    plt.savefig(output_path)  # 
    plt.show()

# jieba 
text = " ".join(df[''].dropna().apply(lambda x: " ".join(jieba.cut(str(x)))))

# 
generate_wordcloud(text, "data/wordcloud.png")

# 
df[''] = pd.to_numeric(df[''], errors='coerce')

#  10 
top_push = df.sort_values(by='', ascending=False).head(10)

# 
plt.figure(figsize=(10,6))
plt.barh(top_push[""], pd.to_numeric(top_push[""], errors="coerce"))
plt.xlabel("")
plt.title("PTT  Top 10")
plt.tight_layout()
plt.savefig("data/top_push.png")
plt.show()
```
